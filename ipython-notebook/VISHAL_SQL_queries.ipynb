## 1 --- List all the directors who directed a 'Comedy' movie in a leap year. (You need to check that the genre is 'Comedy and year is a leap year).Query should return director name, the movie name, and the year.

<h4>To determine whether a year is a leap year,steps are below:</h4>

<ul>
    <li><b>STEP-1:</b> If the year is evenly divisible by 4, go to step 2. Otherwise, go to step 5.</li>
    <li><b>STEP-2:</b> If the year is evenly divisible by 100, go to step 3. Otherwise, go to step 4.</li>
    <li><b>STEP-3:</b> If the year is evenly divisible by 400, go to step 4. Otherwise, go to step 5.</li>
    <li><b>STEP-4:</b> The year is a leap year (it has 366 days).</li>
    <li><b>STEP-5:</b> The year is not a leap year (it has 365 days).</li>
</ul>

Year 1900 is divisible by 4 and 100 but it is not divisible by 400, so it is not a leap year.

```python
%%time


query_5a = """  
                WITH
                OLD_TABLE AS
                (
                SELECT m.MID AS "MID" , p.Gender AS "GENDER", COUNT(MC.MID) AS "COUNT" FROM Movie m 
                JOIN M_Cast MC ON m.MID = MC.MID 
                JOIN Person p ON MC.PID = p.PID
                GROUP BY m.MID,p.Gender
                
                ),
             
                NEW_TABLE AS
                (
                SELECT O.MID ,COUNT(O.MID) AS COUNT_MID , O.Gender FROM OLD_TABLE O
                GROUP BY O.MID
                HAVING COUNT_MID==1 AND O.Gender='Female'
                )


                SELECT m.year AS "Year",n.COUNT_MID AS "Female_Cast_Only_Movies"
                FROM Movie m 
                JOIN NEW_TABLE n ON m.MID=n.MID
                ORDER BY m.year
               
            """
grader_5a(query_5a)
```

## 5.b --- Report for each year the percentage of movies in that year with only female actors, and the total number of movies made that year. For example, one answer will be: 1990 31.81 13522 meaning that in 1990 there were 13,522 movies, and 31.81% had only female actors. You do not need to round your answer.

```python
%%time


query_5b = """  
                WITH
                OLD_TABLE AS
                (
                SELECT m.MID AS "MID" , p.Gender AS "GENDER", COUNT(MC.MID) AS "COUNT" FROM Movie m 
                JOIN M_Cast MC ON m.MID = MC.MID 
                JOIN Person p ON MC.PID = p.PID
                GROUP BY m.MID,p.Gender
                
                ),
             
                NEW_TABLE AS
                (
                SELECT O.MID ,COUNT(O.MID) AS COUNT_MID , O.Gender FROM OLD_TABLE O
                GROUP BY O.MID
                HAVING COUNT_MID==1 AND O.Gender='Female'
                ),


                NEW_TABLE_2 AS
                (
                SELECT m.year AS "year", (n.COUNT_MID) AS "Female" 
                FROM Movie m 
                JOIN NEW_TABLE n ON m.MID=n.MID
                ORDER BY m.year
                ),

                NEW_TABLE_3 AS
                (
                SELECT year,COUNT(title) Total_Movies FROM Movie 
                GROUP BY year 
                ORDER BY year 
                )


                SELECT n.year, ((n.Female*1.0)/nn.Total_Movies) AS "Percentage_Only_Female_Cast_Only_Movies" , nn.Total_Movies FROM NEW_TABLE_2 n
                JOIN NEW_TABLE_3 nn ON n.year=nn.year
               
            """
grader_5b(query_5b)
```

```python
%%time



query7b = """ 

WITH
Years_Unique AS
(
SELECT DISTINCT year FROM Movie 
ORDER BY year
)

SELECT YU.year AS "DECADE_STARTING_YEAR", YU.year+9 AS "DECADE_ENDING_YEAR", COUNT(*) AS "DECADE_MOVIE_COUNT" FROM Years_Unique YU
JOIN Movie m ON (m.year >= DECADE_STARTING_YEAR) AND (m.year <= DECADE_ENDING_YEAR)
GROUP BY DECADE_STARTING_YEAR
ORDER BY DECADE_MOVIE_COUNT DESC

          """
grader_7b(query7b)
```

```python
%%time


query7 = """

WITH
Years_Unique AS
(
SELECT DISTINCT year FROM Movie 
ORDER BY year
),

FINAL_TABLE AS
(
SELECT YU.year AS "DECADE_STARTING_YEAR", YU.year+9 AS "DECADE_ENDING_YEAR", COUNT(*) AS "DECADE_MOVIE_COUNT" FROM Years_Unique YU
JOIN Movie m ON (m.year >= DECADE_STARTING_YEAR) AND (m.year <= DECADE_ENDING_YEAR)
GROUP BY DECADE_STARTING_YEAR
ORDER BY DECADE_MOVIE_COUNT DESC
)

SELECT * FROM FINAL_TABLE
WHERE DECADE_MOVIE_COUNT = (SELECT MAX(DECADE_MOVIE_COUNT) FROM FINAL_TABLE)
"""

grader_7(query7)
```

## Q9 --- The Shahrukh number of an actor is the length of the shortest path between the actor and Shahrukh Khan in the "co-acting" graph.
